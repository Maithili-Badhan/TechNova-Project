# Live Camera Inference

```python
from ultralytics import YOLO
import cv2

model = YOLO("runs/detect/train/weights/best.pt")  # or yolov8s.pt fine-tuned
cap = cv2.VideoCapture(0)  # use RTSP/phone camera URL as needed

CONF_THRES = 0.4

while True:
    ok, frame = cap.read()
    if not ok: break

    results = model(frame, conf=CONF_THRES)[0]
    for box in results.boxes:
        x1,y1,x2,y2 = map(int, box.xyxy[0])
        conf = float(box.conf[0])
        cls  = int(box.cls[0])
        cv2.rectangle(frame, (x1,y1), (x2,y2), (0,255,0), 2)
        cv2.putText(frame, f"billboard {conf:.2f}", (x1, y1-6),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    cv2.imshow("Billboard Detector", frame)
    if cv2.waitKey(1) == 27: break

cap.release(); cv2.destroyAllWindows()
```

# JSON Regulation Checker

```python
import cv2
import json
from ultralytics import YOLO
from datetime import datetime

# Load YOLO model and rules
model = YOLO("best.pt")
with open("regulations.json", "r") as f:
    regulations = json.load(f)

cap = cv2.VideoCapture(0)
CONF_THRES = 0.4

violations_log = []   # store violations for later

def check_policy(detection, regulations):
    rules = regulations["billboard_compliance_rules"]
    violated = []

    # Example: size rule (generic max size)
    for rule in rules:
        if rule["rule_id"] == "size_limits_generic":
            max_w = rule["max_width_m"]
            max_h = rule["max_height_m"]

            # assume 1 pixel = 0.01m (placeholder!)
            px_to_m = 0.01
            x1, y1, x2, y2 = detection["bbox"]
            w = (x2 - x1) * px_to_m
            h = (y2 - y1) * px_to_m

            if w > max_w or h > max_h:
                violated.append(rule["violation_message"])

    return violated

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, conf=CONF_THRES)[0]

    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = float(box.conf[0])
        cls = int(box.cls[0])

        detection_data = {
            "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "bbox": [x1, y1, x2, y2],
            "confidence": conf,
            "class_id": cls,
            "class_name": model.names[cls]
        }

        # Check rules
        violations = check_policy(detection_data, regulations)
        if violations:
            detection_data["violations"] = violations
            violations_log.append(detection_data)  # save violation record

        # Draw results
        if violations:
            color = (0, 0, 255)  # Red for violation
            label = "UNAUTHORISED BILLBOARD"
        else:
            color = (0, 255, 0)  # Green if ok
            label = f"{model.names[cls]} {conf:.2f}"

        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    cv2.imshow("Billboard Compliance Check", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# 🔴 Save violations to file when you stop
if violations_log:
    with open("violations.json", "w") as f:
        json.dump(violations_log, f, indent=4)
    print(f"Saved {len(violations_log)} violations to violations.json")
else:
    print("✅ No violations detected, nothing saved")
```
